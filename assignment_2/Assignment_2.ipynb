### 有三种策略需要考虑：1.最少的站数   2.最短路径  3.最少的换乘
#### 因为上面三种策略的每一个都不仅是一个最有结果，所以还可以细分：最少站数中的最短路径，最小换乘中的最小路径等等

#### 1.最少站点数路线

```python
#最少站点数
def find_path(destination, start, graph, l):
    queue = [[destination]]    #这点技巧是为了加入所有最短路径
    best_path = []
    
    while queue:
        v = queue.pop(0)
        path = v[-1]
        for station in graph[path]:
            if station in l.keys():
                if l[station] == (l[path]-1):
                    new_path = v + [station]
                    queue.append(new_path)
                    if l[station] == 0:
                        best_path.append(new_path)
                    
    solution_path = []
    for path in best_path:
        rev_path = []
        for station in reversed(path):
            rev_path.append(station)
        solution_path.append(rev_path)
    
    return solution_path
```

```python
def find_shortest_edge(graph, start, destination):
    if (start not in graph.keys()) or (destination not in graph.keys()):
        print("{} or {} is not correct station!")
        return None
    else:
        seen = set()
        seen.add(start)    #将第一个点标记为已经看到
        l = {}   #是距离start点最少边数
        l[start] = 0

        queue = [start]   #队列
        while queue:
            v = queue.pop(0)
            for v_station in graph[v]:
                if v_station not in seen:
                    seen.add(v_station)
                    l[v_station] = l[v] + 1
                    queue.append(v_station)
                    if v_station == destination:
                        return l
```

```python
def pretty_print(pathes):
    for index, line in enumerate(pathes):
        print("最少站点可选路线第：{} 条".format(index+1))
        print('->'.join(line))    #join的用法是将jion的列表元素用“->”来连接
    
def least_station_path(graph, start, destination):    #最少乘站可能不只是一条路
    shortest_edge_number = find_shortest_edge(graph, start, destination)
    best_path = find_path(destination, start, graph, shortest_edge_number)
    return best_path
```

```python
import heapq     #用堆取最近的距离
```

```python
def Dijkstra_shortest_path(graph, distance_two_stations, start, destination):
    x = set()  #保存已经确定最短距离的站点
    heap = []
    key = {}    #每一个站点距start点的距离
    Stations = graph.keys()
    for station in Stations:
        key[station] = float("inf")
        
    key[start] = 0
    
    path_dis = {start:0}   #记录搜寻路径中的站点，会有其他不是最短路径上的点
    
    
    for station, distance in key.items():
        heapq.heappush(heap, (distance, station))   #因为station无法比较，用distance最为优先级，并以元组的形式加入堆中
        
    while heap:
        w_distance, w_station = heapq.heappop(heap)
        x.add(w_station)
        for v_station in graph[w_station]:
            
            #判断是找到了destination，并且是在x中的，即可返回
            if destination in x:
                return path_dis
            
            
            if v_station not in x:
                
                for item in heap:   #用于从堆中删除w_connection_station
                    if item[1] == v_station:
                        heap.remove(item)
                        heapq.heapify(heap)    #重新堆化
                        break
                #保持两位小数       
                key[v_station] = round(min(key[v_station], w_distance + distance_two_stations[(w_station, v_station)]), 2)
                heapq.heappush(heap, (key[v_station], v_station))
                
                path_dis[v_station] = key[v_station]
                
```

```python
def find_path(graph, path_distance, distance_two_stations, start, destination):
    up_station = [destination]
    marked = set(up_station)
    
    while True:
        for station in graph[up_station[-1]]:
            if station in path_distance and station not in marked:
                l = round(path_distance[up_station[-1]] - distance_two_stations[(up_station[-1], station)], 2)

                if l == path_distance[station]:
                    up_station.append(station)
                    marked.add(station)
                    
                    if station == start:
                        rev_path = []
                        for station in reversed(up_station):
                            rev_path.append(station)                     
                        return rev_path
                     
                    break
```

```python
def shortest_distance_path(graph, distance_two_stations, start, destination):
    path_distance = Dijkstra_shortest_path(graph, distance_two_stations, start, destination)
    best_path = find_path(graph, path_distance, distance_two_stations, start, destination)
    return best_path
```
